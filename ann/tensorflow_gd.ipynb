```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = tf.Variable(100.0)
```

```python
cost = u*u-4*u+4
```

```python
train_op = tf.train.GradientDescentOptimizer(0.3).minimize(cost)
```

```python
init = tf.global_variables_initializer()
u_cal = []
cost_cal = []
with tf.Session() as session:
    session.run(init)
    for i in xrange(30):
        session.run(train_op)
        u_cal.append(u.eval())
        cost_cal.append(cost.eval())
```

```python
plt.plot(u_cal)
plt.show()
```

```python
plt.plot(cost_cal)
plt.show()
```
