```python
# This model finds the best fit quadratic polynomial for the data.
# We can use multiple linear regression solving method (Linear regression having multiple variables.),
# as, in y = ax^2 + bx + c : y is linear to a,b and c. 
```

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
```

```python
for line in open("../files/regression/data_poly.csv"):
    line = line.split(",")
    x = float(line[0])
    y = float(line[1])
    X.append([1,x,x*x])
    Y.append(y)
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X[:,1],Y)
plt.show()
```

```python
# Calculating weight
W = np.linalg.solve(X.T.dot(X), X.T.dot(Y))
# We write the code as Y = Wtranspose X. 
# But here X is an NXD matrix where each sample is row. It is easier to do XW
```

```python
# We calculate Yhat and the R^2 value.
```

```python
Yhat = np.dot(X,W)

# Compute R squared.
SSres = Y-Yhat
SStot = Y-Y.mean()

r_sq = 1 - SSres.dot(SSres) / (SStot.dot(SStot))

print "r_squared =", r_sq 
```

```python
# Lets plot the curve
plt.scatter(X[:,1],Y)
plt.plot(sorted(X[:,1]),sorted(Yhat))
plt.show()
```
