# Demo project to create a dataset of Donut shape

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = 5*np.random.randn(1000,2)
data2 = np.random.randn(1000,2)
```

```python
plt.scatter(data1[:,0],data1[:,1], c='b')
plt.scatter(data2[:,0],data2[:,1], c='r')
plt.show()
```

```python
# Remove all data points having norm > 8
data3 = data1[np.linalg.norm(data1, axis=1) > 8]
```

```python
plt.scatter(data3[:,0],data3[:,1], c='b')
plt.scatter(data2[:,0],data2[:,1], c='r')
plt.show()
```

```python
# The donut shape is created
```

```python
N = 1000
D = 2

R_inner = 5
R_outer = 10

# distance from origin is radius + random normal
# angle theta is uniformly distributed between (0, 2pi)
R1 = np.random.randn(N//2) + R_inner
theta = 2*np.pi*np.random.random(N//2)
X_inner = np.concatenate([[R1 * np.cos(theta)], [R1 * np.sin(theta)]]).T

R2 = np.random.randn(N//2) + R_outer
theta = 2*np.pi*np.random.random(N//2)
X_outer = np.concatenate([[R2 * np.cos(theta)], [R2 * np.sin(theta)]]).T

X = np.concatenate([ X_inner, X_outer ])
T = np.array([0]*(N//2) + [1]*(N//2)) # labels: first 50 are 0, last 50 are 1

plt.scatter(X[:,0], X[:,1], c=T)
plt.show()
```
